```python
with open(r'/workspaces/pyskl/workspace/results/real_time/all_uncertainty.pkl', 'rb') as f:
# with open(r'/workspaces/pyskl/workspace/results/unseen/uncertainty.pkl', 'rb') as f:
    context = pk.load(f)
    
uncertainty = context['uncertainty_list']
print(len(uncertainty))
# np.random.seed(5)
idx = 0
for item in range(len(uncertainty)-1, -1, -1):
    uncertainty[item] = uncertainty[item]
    if uncertainty[item] > 0.2:
        idx += 1
print(f"num: {idx}")
        

# plt.figure(figsize=(12, 8))  #设置画布的大小
# sns.set_palette("hls")       #设置所有图的颜色，使用hls色彩空间
plt.style.use('default')
palette = plt.get_cmap('tab20b')
print(len(palette.colors))

# sns.histplot(prediction,color="steelblue",bins=50)
ax = sns.lineplot(uncertainty[::2][:60])
# ax.set(yscale="log")
plt.xlabel('frames',fontsize=15)           #添加x轴标签，并改变字体
plt.ylabel('Uncertainty',fontsize=15)   #添加y轴变浅，并改变字体
plt.grid(linestyle='-')   #添加网格线
plt.xticks(fontsize=12)   #改变x轴字体大小
plt.yticks(fontsize=12)   #改变y轴字体大小
sns.despine(ax=None, top=True, right=True, left=True,bottom=True)    #将图像的框框删掉
plt.show()
plt.savefig(r'/workspaces/pyskl/workspace/notebooks/real_time_uncertianty.svg')
```
